# ***2-3 득실 그래프***
팀별 득점과 실점을 계산해 별도의 그래프를 그린 뒤 최다 득점, 최소 실점팀 만을 다른 색으로 표시하였다.
(실점 결과가 200인 경우는 강등으로 1부리그에 존재하지 않을때 이다)


***결과 분석***

이전 분석에서 보았듯 2010년즈음을 기준으로 최다 득점팀이 맨유, 맨시티 위주로 갈림을 볼 수 있다.

```python
big6 = ['Man United', 'Arsenal', 'Liverpool', 'Chelsea', 'Tottenham', 'Man City']

season_max_goals = {}
season_min_conceded = {}

#팀별 득점, 실점을 계산한다.
for season in dfDroped['Season'].unique():
    season_data = dfDroped[dfDroped['Season'] == season]
    season_max_goals[season] = {}
    season_min_conceded[season] = {}
    for club in big6:
        home_list = season_data['HomeTeam'].to_list()
        away_list = season_data['AwayTeam'].to_list()

        if (club not in home_list) and (club not in away_list):
          season_max_goals[season][club] = 0
          season_min_conceded[season][club] = 200
          continue

        home_club_data = season_data[season_data['HomeTeam'] == club]
        away_club_data = season_data[season_data['AwayTeam'] == club]


        goals_scored = home_club_data['FTHG'].sum() + away_club_data['FTAG'].sum()
        goals_conceded = home_club_data['FTAG'].sum() + away_club_data['FTHG'].sum()

        season_max_goals[season][club] = goals_scored
        season_min_conceded[season][club] = goals_conceded


seasons = dfDroped['Season'].unique()

colors = ['gray' for _ in big6]
fig, ax = plt.subplots(1, len(seasons), figsize=(40, 6), sharey=True)
fig.suptitle('Max goal team per season')

for i, season in enumerate(seasons):
    clubs = list(season_max_goals[season].keys())
    goals_max = list(season_max_goals[season].values())

    max_goals_club = max(season_max_goals[season], key=season_max_goals[season].get)

    colors = ['yellow' if club == max_goals_club else 'gray' for club in clubs]
    bars = ax[i].barh(clubs, goals_max, color=colors)

    for bar, goals in zip(bars, goals_max):
        ax[i].text(goals, bar.get_y() + bar.get_height() / 2, str(goals), ha='right', va='center', color='black')

    ax[i].barh(clubs, goals_max, color=colors)
    ax[i].set_xlabel('Goals')
    ax[i].set_title(season)

fig2, ax2 = plt.subplots(1, len(seasons), figsize=(40, 6), sharey=True)
fig2.suptitle('Min conceded goals per season')
for i, season in enumerate(seasons):
    clubs = list(season_min_conceded[season].keys())
    min_conceded = list(season_min_conceded[season].values())


    min_conceded_club = min(season_min_conceded[season], key=season_min_conceded[season].get)

    colors = ['blue' if club == min_conceded_club else 'gray' for club in clubs]
    bars = ax2[i].barh(clubs, goals_max, color=colors)

    for bar, goals in zip(bars, min_conceded):
        ax2[i].text(goals, bar.get_y() + bar.get_height() / 2, str(goals), ha='right', va='center', color='white')

    ax2[i].barh(clubs, min_conceded, color=colors)
    ax2[i].set_xlabel('Goals')
    ax2[i].set_title(season)
```

# ***2-4 승률***
맨유의 승률이 가장 높고 토트넘의 승률이 가장 낮음을 확인할 수 있다.

해외축구 팬덤 사이에서 토트넘이 빅6에 자리할 자격이 있나에 대해 말이 나오곤 한다.

지금까지의 분석 결과를 보았을때 우승경험이 없고, 순위와 득실면에서 우수한 성적을 거둔 시즌이 몇 없으며, 빅6 팀중 승률이 가장 낮은것으로 보아 과연 토트넘이 빅6의 자격이 있는가라는 주장에 대한 근거가 충분한 것으로 보인다.

```python
df_copy = df_original.copy()
df_copy.tail()

big6 = ['Man United', 'Man City', 'Liverpool', 'Chelsea', 'Arsenal', 'Tottenham']

team_stats = {team: {'games': 0, 'wins': 0} for team in big6}

for _, row in df_copy.iterrows():
    home_team = row['HomeTeam']
    away_team = row['AwayTeam']
    match_result = row['FTR']

    if home_team in big6:
        team_stats[home_team]['games'] += 1
        if match_result == 'H':
            team_stats[home_team]['wins'] += 1

    if away_team in big6:
        team_stats[away_team]['games'] += 1
        if match_result == 'A':
            team_stats[away_team]['wins'] += 1

win_rates = {team: 0.0 for team in big6}

for team in big6:
    games = team_stats[team]['games']
    wins = team_stats[team]['wins']
    win_rate = wins / games if games > 0 else 0
    win_rate *= 100
    win_rates[team] = win_rate

team_colors = {
    'Man United': 'red',
    'Arsenal': 'orange',
    'Chelsea': 'blue',
    'Man City': 'cyan',
    'Tottenham': 'green',
    'Liverpool': 'darkred'
}

plt.figure(figsize=(10, 6))
plt.grid(True)

teams = win_rates.keys()
win_rates_val = win_rates.values()

bar = plt.bar(teams, win_rates_val, color=[team_colors[team] for team in big6])

plt.ylim(0, 100)
plt.xlabel('Team')
plt.ylabel('Win rate')
plt.title('Win rate of Big6 teams in PL from 93-94 to 21-22')

for rect in bar:
    height = rect.get_height()
    plt.text(rect.get_x() + rect.get_width()/2.0, height, '%.1f' % height, ha='center', va='bottom', size = 12)

plt.tight_layout()
plt.show()
```

# ***2-5 빅6 팀별 승률***
빅6 팀마다 각각 다른 5개의 팀과의 경기에서의 승률을 그래프로 나타내었다.

***결과분석***

맨유의 전체적인 승률이 높음을 볼 수 있다. 전체적으로 토트넘과 맨시티에 대한 승률이 비교적 높게 나옴을 확인할 수 있다.

2010년부터 상승기에 접어든 맨시티 이지만 2010년 이전까지는 부족한 성적을 거두었다고 추측해볼 수 있다.

토트넘의 전체적인 승률은 낮지만 유독 맨시티전에선 절반이 넘는 승을 거두었음을 볼 수 있다,

토트넘이 맨시티의 천적으로 알려져 있는데 통계적으로 정말 천적이 맞음을 확인할 수 있다.

```python
import copy

df_copy = df_original.copy()
df_copy.tail()

big6 = ['Man United', 'Man City', 'Liverpool', 'Chelsea', 'Arsenal', 'Tottenham']

team_stats = {team: {opponent: {'games': 0, 'wins': 0} for opponent in big6 if opponent != team} for team in big6}

for _, row in df_copy.iterrows():
    home_team = row['HomeTeam']
    away_team = row['AwayTeam']
    match_result = row['FTR']

    if home_team in big6 and away_team in big6:
        team_stats[home_team][away_team]['games'] += 1
        team_stats[away_team][home_team]['games'] += 1

        if match_result == 'H':
            team_stats[home_team][away_team]['wins'] += 1
        elif match_result == 'A':
            team_stats[away_team][home_team]['wins'] += 1

team_colors = {
    'Man United': 'red',
    'Arsenal': 'orange',
    'Chelsea': 'blue',
    'Man City': 'cyan',
    'Tottenham': 'green',
    'Liverpool': 'darkred'
}
win_rates = {team: {} for team in big6}
for team in big6:
    for opponent in team_stats[team]:
        games = team_stats[team][opponent]['games']
        wins = team_stats[team][opponent]['wins']
        win_rate = (wins / games) if games > 0 else 0
        win_rate *= 100
        win_rates[team][opponent] = win_rate

fig, ax = plt.subplots(2, 3, figsize=(20, 10))
fig.suptitle('Win rate of Big6 against other Big6 teams')

i = 0
j = 0
for team in win_rates.keys():
  teams = win_rates[team].keys()
  win_rates_val = win_rates[team].values()

  colors = [team_colors[opponent] for opponent in teams]

  bar = ax[i][j].bar(teams, win_rates_val, color=colors)
  ax[i][j].set_ylim(0,100)
  ax[i][j].set_title("%s" % team)
  ax[i][j].grid(True)
  for rect in bar:
    height = rect.get_height()
    ax[i][j].text(rect.get_x() + rect.get_width()/2.0, height, '%.1f' % height, ha='center', va='bottom', size = 12)
  j += 1
  if (j == 3):
    i = 1
    j = 0
```

# ***3) 머신러닝***

FTR(경기 결과)를 예측하는 모델을 만들고 학습시켜 예측을 해볼 것 이다.
경기 결과와 경기 세부 지표들만이 필요하기때문에 경기 결과에 영향이 전혀 없거나 아주 작은 피쳐들과 FTR(종속변수)를 제외한 피쳐를 독립변수로 할당한 후
학습, 예측, 평가 과정을 진행할 것이다.

또한 독립변수 피쳐들의 중요도 분석을 진행하여 경기 결과에 큰 영향을 미치는 지표들을 살펴볼것이다.

***결과 분석***

홈팀 승리, 어웨이팀 승리에 대해선 준수한 결과를 보이지만 비겼을 경우엔 상당히 낮은 평가 지표를 확인할 수 있다.

원인을 추측해볼 수 있는데,

비긴 경기의 경우 승,패가 정해진 경기와 달리 치열한 접전이 벌어지거나 두 팀이 소극적으로 경기에 임하거나 때로는 운 등의 예측하기 쉽지 않은 경향을 보인다고 생각할 수 있다.

피쳐 중요도 분석의 경우 HTR(Half Time Result, 하프타임 결과)가 큰 영향을 미침을 알 수 있다,

하프타임에 지고있는 팀의 입장에선 누적된 결과, 심리적요인과 후반으로 갈 수록 쌓이는 육체적 피로도로 인해 경기를 뒤집기 쉽지 않다고 추측해볼 수 있다.


   

```python
from sklearn.metrics import classification_report
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import roc_curve, roc_auc_score

df_half = df_original.copy().dropna()

df_half = df_half.drop(axis = 1, columns = ['Referee', 'HomeTeam', 'AwayTeam', 'Season', 'DateTime'])
#홈팀이 이기면 2, 어웨이팀이 이기면 1, 비기면 0
df_half['FTR'].replace(['H', 'A', 'D'], [2, 1, 0], inplace = True)
df_half['HTR'].replace(['H', 'A', 'D'], [2, 1, 0], inplace = True)

# 하프타임 결과 피처가 있기에 하프타임 골수는 종속변수에 포함하지 않는다.
# FTAG, FTHG 피처로 FTR을 바로 예측할 수 있으므로 종속변수에 포함하지 않는다.
X = df_half[df_half.columns.difference(['FTR', 'FTAG', 'FTHG','HTAG', 'HTHG'])]
y = df_half['FTR']


X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.3, random_state = 42)

model = LogisticRegression(max_iter = 1000, multi_class = 'multinomial')

model.fit(X_train, y_train)

y_pred = model.predict(X_test)
y_pred_prob = model.predict_proba(X_test)
print(classification_report(y_test, y_pred))


feature_importance = abs(model.coef_[0])
features = X_train.columns

sorted_idx = feature_importance.argsort()

plt.figure(figsize=(10, 6))
bars = plt.barh(range(len(sorted_idx)), feature_importance[sorted_idx], align='center')
plt.yticks(range(len(sorted_idx)), [features[i] for i in sorted_idx])
plt.xlabel('Feature Importance')
plt.ylabel('Feature')
plt.title('Feature Importance in Logistic Regression Model')
plt.xlim(0,0.4)
for bar in bars:
    plt.text(bar.get_width() + 0.01, bar.get_y() + bar.get_height()/2,
             f'{bar.get_width():.4f}',
             va='center', ha='left', color='black')

plt.show()
```

# ***4) 결론***

Kaggle의 93-94 to 21-22 PL data를 탐색적 분석과 머신러닝을 이용해 분석해보았다.
제공된 데이터가 경기, 팀에 대한 데이터이었기 때문에 선수에 대한 분석(최다 득점자, 최소 실점자, 촤다 출장자 등)은 할 수 없었던점이 아쉬웠다.

승강등 시스템과 데이터 분석의 복잡성의 이유로 빅6 팀 중심적 분석을 하였는데
전체 20개 팀에 대한 분석을 한다면 머신러닝을 중심으로 피처들의 중요도를 살펴보며 축구 경기의 경향을 위주로 분석할 수 있으며 강등 팀, 우승 팀의 공통점 등과 같은 분석도 가능하다고 생각한다.

***소감***

예제 책에 나와있는 날씨, 인구 등의 공공데이터는 다른 학우들과 중복, 관심이 없는 등의 이유로 배제하고 분석 자체에 재미를 느낄만한 데이터를 찾았는데, 그렇게 선택하게 된 데이터가 프리미어리그 데이터였다. 데이터 자체에 재미를 느껴 9장 데이터 분석 파이프 라인의 내용 외에도 다른 분석들을 생각해보고 적용하는 과정이 재미있었다.

데이터를 다루며 원하는 결과를 얻기 위해 데이터 프레임을 조작하고 이것저것 시도해보며 수업시간에 배운 내용들을 토대로 더 많은것을 배울 수 있었다.

머신러닝 파트에서 FTR을 예측할때 FTHG, FTAG(풀타임 홈, 어웨이 골)을 독립변수에 포함시키는 바람에 모든 평가지표가 1.00 이 나와 오랜시간 고민을 했었다. 결국에는 원인을 찾아 올바른 독립변수를 할당하였지만 이 경험으로 머신러닝과 데이터의 특성에 대한 이해도가 살짝이나마 올라갔다고 생각한다.
